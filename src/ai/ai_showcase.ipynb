```python
from randomForest import RandomForest
import numpy as np
from confMatr import ConfusionMatrix
from ai import get_edited_fight
```

```python
train_data = []
for fight_id in range(12003, 12050):
    fight1, fight2 = get_edited_fight(fight_id)
    train_data.append(fight1)
    train_data.append(fight2)


test_data = []
for fight_id in range(12000, 12003):
    fight1, fight2 = get_edited_fight(fight_id)
    test_data.append(fight1)
    test_data.append(fight2)

print(train_data)
test_data
```

```python
cfm = ConfusionMatrix()
randomForest = RandomForest(10)
randomForest.fit(np.array(train_data))

for fight in np.array(test_data):
    cfm.evaluate(randomForest.predict(fight), fight[-1])
cfm.print_all()
```
